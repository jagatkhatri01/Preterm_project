```python
corr = num_data.corr('pearson')
def high_corr_counts(corr_df, thresh):
    absc = corr_df.abs()
    n = absc.shape[0]
    mask = np.triu(np.ones((n,n), dtype=bool), k=1)
    return (absc.values[mask] > thresh).sum()

for t in [0.95, 0.90, 0.85, 0.80]:
    print(f"Pairs with |corr|>{t}: {high_corr_counts(corr, t)}")
```

```python
num_cols = [c for c in data_cleaned.columns 
            if c != 'preterm' and data_cleaned[c].dtype != 'object']
num_data = data_cleaned[num_cols]
num_data_clean = num_data.loc[:, num_data.nunique() > 1]
print(f"Original numeric columns: {len(num_data.columns)}")
print(f"Columns for clustering: {len(num_data_clean.columns)}")
corr = num_data_clean.corr('pearson').abs()
dist = 1 - corr
condensed_dist = squareform(dist.values, checks=False)
print("Condensed distance length:", len(condensed_dist))
Z = linkage(condensed_dist, method='average')
```

```python
plt.figure(figsize=(15, 6))
plt.title("Feature Clustering Dendrogram")
dendrogram(Z, labels=num_data_clean.columns, leaf_rotation=90, leaf_font_size=8)
plt.tight_layout()
plt.show()
```

```python
# Choose correlation threshold
corr_threshold = 0.80  # you can tune: 0.95=very conservative, 0.85=aggressive
cluster_threshold = 1 - corr_threshold

# Assign cluster IDs to each feature
clusters = fcluster(Z, t=cluster_threshold, criterion='distance')
feat2cluster = pd.Series(clusters, index=num_data_clean.columns)

# See cluster sizes
print(feat2cluster.value_counts().sort_index())
```

```python
from sklearn.feature_selection import mutual_info_classif

# Impute missing values for numeric features
X_for_mi = num_data_clean.copy()  # only continuous features
y = data_cleaned['preterm']

# Calculate MI for all continuous numeric features
mi = mutual_info_classif(X_for_mi, y, discrete_features=False, random_state=0)
mi_s = pd.Series(mi, index=X_for_mi.columns).sort_values(ascending=False)
print(len(mi_s))
print(mi_s.head(10))  # top 10 most informative features
```

```python
selected_reps = []

# Go cluster by cluster
for cid, members in feat2cluster.groupby(feat2cluster).groups.items():
    # Pick the member with highest MI
    best_feature = mi_s[members].idxmax()
    selected_reps.append(best_feature)

print(f"Selected representatives: {len(selected_reps)}")
print(selected_reps)
print(selected_reps[:20])
```

```python
import pandas as pd

# Load your data


# Create summary for exactly 12 columns
def create_sample_summary(dataframe, num_columns=100):
    # Select first 'num_columns' columns
    sample_df = dataframe.iloc[:, :num_columns]
    
    summary = pd.DataFrame({
        'name of column': sample_df.columns,
        'types': sample_df.dtypes.values,
        'unique_data': sample_df.nunique().values,
        'missing value': sample_df.isnull().sum().values,
        'missing percentage': (sample_df.isnull().sum() / len(sample_df) * 100).round(6),
        'duplicated': [sample_df.duplicated().sum()] * len(sample_df.columns)  # Same for all columns
    })
    return summary

# Create summary for 12 columns
sample_summary = create_sample_summary(data_cleaned, 100)
sample_summary
```
