```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from scipy.spatial.distance import squareform
```

```python
data = pd.read_csv('Dataset/natality2024ps.csv')
```

```python
same_value_cols = [col for col in data.columns if data[col].nunique()==1]
print("Columns with same value in all rows")
print(same_value_cols)
```

```python
missing_80_cols = [col for col in data.columns if data[col].isnull().mean()>0.9]
print("Columns with more than 80% missing values: ")
print(missing_80_cols)
```

```python
cols_to_drop = list(set(same_value_cols + missing_80_cols))
data_cleaned = data.drop(columns=cols_to_drop)
print(cols_to_drop)

remaining = [col for col in cols_to_drop if col in data_cleaned.columns]
print("Columns still present after deletion (should be empty):")
print(remaining)

print("Original Shape", data.shape)
print("New Shape", data_cleaned.shape)
```

```python
data_cleaned['preterm'] = (data_cleaned['oegest_comb']<37).astype(int)
```

```python
data_cleaned.drop(columns=['gestrec10',
 'gestrec3',
 'oegest_comb',
 'oegest_r10',
 'oegest_r3'], inplace=True)
```

```python
data_cleaned.shape
```

```python
data_cleaned = data_cleaned.dropna(subset=['rcnty_pop'])
```

```python
num_cols = data_cleaned.select_dtypes(include=['number']).columns.tolist()
num_cols.remove('preterm')  # remove target
cat_cols = data_cleaned.select_dtypes(include=['object']).columns.tolist()

# Numeric features
X_num = data_cleaned[num_cols].copy()

# Categorical features → ordinal encoding
X_cat = data_cleaned[cat_cols].copy()
X_cat_encoded = X_cat.apply(lambda col: col.astype('category').cat.codes)

# Combine numeric + encoded categorical
X_rfe = pd.concat([X_num, X_cat_encoded], axis=1)
y_rfe = data_cleaned['preterm']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE

# Base model
rf = RandomForestClassifier(n_estimators=200, random_state=0, n_jobs=-1)

# RFE selector — choose number of features you want
n_features_to_select = 60
rfe_selector = RFE(estimator=rf, n_features_to_select=n_features_to_select, step=10, verbose=1)

# Fit RFE
rfe_selector.fit(X_rfe, y_rfe)

# Get selected feature names
selected_features_rfe = X_rfe.columns[rfe_selector.support_].tolist()
print(f"Top {n_features_to_select} features selected by RFE:")
print(selected_features_rfe)
```
